```python
from transformers import pipeline

generator  = pipeline("text-generation", model="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B", device_map="auto")

result=generator(
    "一個農場有雞跟兔，共有5個頭16隻腳，請問雞跟兔各有幾隻?請用繁體中文回答",
    truncation=True,
    max_length=30000,
    num_return_sequences=1,
)

print(result)
```
